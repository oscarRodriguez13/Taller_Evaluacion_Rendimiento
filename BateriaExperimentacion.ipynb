# Analisis de los algoritmos

Empezamos a importar el dataframe del csv

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_clasico_1 = pd.read_csv("CSV_MM_clasico_Pc1.csv")
df_clasico_2 = pd.read_csv("CSV_MM_clasico_Pc2.csv")
df_transpuesta_1 = pd.read_csv("CSV_MM_transpuesta_Pc1.csv")
df_transpuesta_2 = pd.read_csv("CSV_MM_transpuesta_Pc2.csv")
```

# Algoritmo Clasico con el PC1

```python
# Calcular el promedio para cada columna
promedios_columnas = df_clasico_1[["200, 1 hilo", "200, 2 hilos", "200, 4 hilos", "200, 6 hilos", "200, 8 hilos", "400, 1 hilo", "400, 2 hilos", "400, 4 hilos", "400, 6 hilos", "400, 8 hilos", "800, 1 hilo", "800, 2 hilos", "800, 4 hilos", "800, 6 hilos", "800, 8 hilos", "1000, 1 hilo", "1000, 2 hilos", "1000, 4 hilos", "1000, 6 hilos", "1000, 8 hilos", "1500, 1 hilo", "1500, 2 hilos", "1500, 4 hilos", "1500, 6 hilos", "1500, 8 hilos", "2000, 1 hilo", "2000, 2 hilos", "2000, 4 hilos", "2000, 6 hilos", "2000, 8 hilos", "2500, 1 hilo", "2500, 2 hilos", "2500, 4 hilos", "2500, 6 hilos", "2500, 8 hilos", "4000, 1 hilo", "4000, 2 hilos", "4000, 4 hilos", "4000, 6 hilos", "4000, 8 hilos"]].mean()


# Convertir los promedios a un dataframe para mejor visualización
tabla_promedios_columnas_clasico1 = pd.DataFrame(promedios_columnas, columns=['Promedio']).reset_index()
tabla_promedios_columnas_clasico1.columns = ['Hilos', 'Promedio']


print("\nTabla de Promedios por Columna:\n", tabla_promedios_columnas_clasico1)
```

```python
# Separar los hilos y convertirlos en listas separadas
hilos = tabla_promedios_columnas_clasico1['Hilos'].str.split(', ', expand=True)
numeros_hilos = hilos[0]
num_hilos = hilos[1]

# Obtener los promedios
promedios = tabla_promedios_columnas_clasico1['Promedio']

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))

# Dibujar una línea para cada número de hilos
for num_hilo in num_hilos.unique():
    plt.plot(numeros_hilos[num_hilos == num_hilo], promedios[num_hilos == num_hilo], marker='o', label=f'{num_hilo} hilos')

# Añadir etiquetas y título
plt.xlabel('Número')
plt.ylabel('Promedio')
plt.title('Promedio de Tiempo por Número de Hilos')
plt.xticks(rotation=45)

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.tight_layout()
plt.grid(True)  # Agregar rejilla
plt.show()
```
